```python
import pandas as pd
```

```python
from fitting_with_ligthgbm import readCsvFiles, ligthData, train_test, makePred
```

```python
x_to_keep = ["TaxRateArea", "totBuildingDataLines", "EffectiveYearBuilt", "SQFTmain", 
                "LandValue", "Bedrooms", "Bathrooms", "zip_class"]
to_keep = x_to_keep + ["TotalValue", "rowID"]
```

```python
train, test, val = readCsvFiles()
```

```python
x = train[x_to_keep]
y = train["log_total_value"]

join_val_test = pd.concat([test,val], axis=0)
x_test = join_val_test[x_to_keep]
y_test = join_val_test["log_total_value"]
```

```python
train_data, test_data = ligthData(x, y, x_test, y_test)
```

```python
model = train_test(train_data, test_data)
```

```python
pred, residual = makePred(model, x, y, x_test, y_test)
```

```python
train["residual"] = residual["train"]
train["log_total_value_pred"] = pred["train"]
```

```python
join_val_test["residual"] = residual["test"]
join_val_test["log_total_value_pred"] = pred["test"]
```

```python
data = pd.concat([train,join_val_test], axis=0)
```

```python
to_keep = ['TotalValue', 'rowID', 'log_total_value', 'residual', 'log_total_value_pred']
```

```python
data = data[to_keep]
```
